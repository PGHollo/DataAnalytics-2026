### Exercise 2.E Using functions for analyzing data
This notebook analyzes a small sales dataset using Python functions.
Each function takes the full dataset as its first argument.

```
sales_data = [
    ('Laptop', 'East', 1200.00),
    ('Phone', 'West', 850.00),
    ('Tablet', 'East', 400.00),
    ('Monitor', 'South', 650.00),
    ('Keyboard', 'West', 75.00),
    ('Laptop', 'South', 1150.00),
    ('Phone', 'East', 900.00),
    ('Tablet', 'West', 425.00),
    ('Monitor', 'East', 680.00),
    ('Keyboard', 'South', 80.00),
    ('Laptop', 'West', 1300.00),
    ('Phone', 'West', 820.00),
]
```

## Lab 1. Total sales by region

```
def region_total(data, region):
    total = 0

    for record in data:
        record_region = record[1]
        amount = record[2]

        if record_region == region:
            total += amount

    return total
```

```
print(region_total(sales_data, 'East'))
print(region_total(sales_data, 'West'))
print(region_total(sales_data, 'South'))
```

## Lab 2. Count of sales by region

```
def region_count(data, region):
    count = 0

    for record in data:
        record_region = record[1]

        if record_region == region:
            count += 1

    return count
```

```
print(region_count(sales_data, 'East'))
print(region_count(sales_data, 'West'))
print(region_count(sales_data, 'South'))
```

## Lab 3. Average sale amount by region

```
def region_average(data, region):
    total = region_total(data, region)
    count = region_count(data, region)

    if count == 0:
        return 0
    else:
        return total / count
```

```
print(region_average(sales_data, 'East'))
print(region_average(sales_data, 'West'))
print(region_average(sales_data, 'South'))
print(region_average(sales_data, 'North'))
```

## Lab 4. Filter by sale amount

```
def sales_above(data, threshold):
    results = []

    for record in data:
        amount = record[2]

        if amount > threshold:
            results.append(record)

    return results

def sales_below(data, threshold):
    results = []

    for record in data:
        amount = record[2]

        if amount < threshold:
            results.append(record)

    return results
```

```
above_500 = sales_above(sales_data, 500)

for record in above_500:
    print(record)
```

```
below_500 = sales_below(sales_data, 500)

for record in below_500:
    print(record)
```

## Lab 5. Top and bottom sale

```
def top_sale(data):
    top_record = data[0]

    for record in data:
        if record[2] > top_record[2]:
            top_record = record

    return top_record

def bottom_sale(data):
    bottom_record = data[0]

    for record in data:
        if record[2] < bottom_record[2]:
            bottom_record = record

    return bottom_record

print(top_sale(sales_data))
print(bottom_sale(sales_data))
```

The top sale was a Laptop in the West region for $1,300.00. The bottom sale was a Keyboard in the West region for $75.00.

## Lab 6. Apply a discount

```
def apply_discount(data, discount_pct):
    new_list = []

    for record in data:
        product = record[0]
        region = record[1]
        amount = record[2]
        new_amount = amount * (1 - discount_pct)
        new_record = (product, region, new_amount)
        new_list.append(new_record)

    return new_list

discounted_sales = apply_discount(sales_data, 0.15)

print('Before discount:')
print(sales_data[0])
print(sales_data[1])
print(sales_data[2])

print('After discount:')
print(discounted_sales[0])
print(discounted_sales[1])
print(discounted_sales[2])
```

## Lab 7. Unique regions

```
def unique_regions(data):
    regions = []

    for record in data:
        region = record[1]

        if region not in regions:
            regions.append(region)

    regions.sort()
    return regions

print(unique_regions(sales_data))
```

This function is useful because it shows what groups exist in the data. In a longer analysis program, I could use this pattern to find unique products, regions, categories, or customer types before making summary reports.

## Lab 8. Region summary

```
def region_summary(data, region):
    region_records = []

    for record in data:
        if record[1] == region:
            region_records.append(record)

    print(f'=== {region} Region ===')
    print(f'Records: {region_count(data, region)}')
    print(f'Total Sales: ${region_total(data, region):,.2f}')
    print(f'Average Sale: ${region_average(data, region):,.2f}')

    if len(region_records) > 0:
        best_sale = top_sale(region_records)
        print(f'Top Sale: {best_sale[0]} — ${best_sale[2]:,.2f}')

region_summary(sales_data, 'East')
```

## Lab 9. Full summary

```
regions = unique_regions(sales_data)

for region in regions:
    region_summary(sales_data, region)
    print()
```

I learned how to use functions to total, count, average, filter, and summarize data. I also learned that writing smaller functions makes the final report easier to build.
