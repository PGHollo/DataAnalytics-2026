# Exercise 2.A Using built-in functions
## Lab 2. Parsing strings – part codes

```
def get_supplier_code(part_code):
    colon_position = part_code.index(':')
    return part_code[:colon_position]

def get_product_num(part_code):
    colon_position = part_code.index(':')
    dash_position = part_code.index('-')
    return part_code[colon_position + 1:dash_position]

def get_size(part_code):
    dash_position = part_code.index('-')
    return part_code[dash_position + 1:]
```

```
code1 = 'ACME:123-L'
code2 = 'DI:12345-M'
code3 = 'ACE:1-12'
```

```
supplier = get_supplier_code(code1)
part_num = get_product_num(code1)
size = get_size(code1)
print(f'For {code1}, the supplier is {supplier}. Part number {part_num} is available in size {size}.')
```

```
supplier = get_supplier_code(code2)
part_num = get_product_num(code2)
size = get_size(code2)
print(f'For {code2}, the supplier is {supplier}. Part number {part_num} is available in size {size}.')
```

```
supplier = get_supplier_code(code3)
part_num = get_product_num(code3)
size = get_size(code3)
print(f'For {code3}, the supplier is {supplier}. Part number {part_num} is available in size {size}.')
```

```
part_codes = [code1, code2, code3]

for code in part_codes:
    supplier = get_supplier_code(code)
    part_num = get_product_num(code)
    size = get_size(code)
    print(f'For {code}, the supplier is {supplier}. Part number {part_num} is available in size {size}.')
```
