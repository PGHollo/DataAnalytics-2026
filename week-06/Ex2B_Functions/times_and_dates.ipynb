Lab 2: Working with datetime functions.

```python
import datetime
```

# Today

```python
today = datetime.datetime.now()
```

```python
day_name = today.strftime('%A')
month_name = today.strftime('%B')
day_num = today.day
year_num = today.year

print(f'Today is {day_name}, {month_name} {day_num}, {year_num}')
```

# My birthday

```python

birthday = datetime.date(1997, 6, 16)
```

```python
birthday_text = birthday.strftime('%m/%d/%Y')
print(f'My birthday is {birthday_text}')
```

# Next 90 Days

```python
ninety_d = today + datetime.timedelta(days=90)
```

```python
month_name = ninety_d.strftime('%B')
print(f'90 days from today is {month_name} {ninety_d.day}, {ninety_d.year}')
```

# Time for Dinner

```python
dinner_time = datetime.time(19, 30)
```

```python
dinner_text = dinner_time.strftime('%I:%M %p')
print(f"Let's meet for dinner at {dinner_text}")
```
