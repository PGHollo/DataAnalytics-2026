# Exercise 5.A NumPy Fundamentals

```
import numpy as np
```

## Create a NumPy array from a list

```
fruits = ['apple', 'banana', 'cherry', 'date']
```

```
arr_1 = np.array(fruits)
print(arr_1)
```

```
print(arr_1.dtype)
# U means Unicode string. The number shows the longest string length.
```

## Create and transpose a 2D array

```
arr_2 = np.array([(10, 11, 12), (13, 14, 15)])
print(arr_2)
```

```
print(arr_2.T)
```

```
print(arr_2)
# arr_2.T does not permanently change arr_2.
```

## Practice selecting array values

```
arr_3 = np.array([(1, 3, 5), (2, 4, 6)])
print(arr_3)
```

```
arr_4 = np.array([arr_3[0, 2], arr_3[1, 1]])
print(arr_4)
```

```
arr_5 = np.ones((2, 2, 2), dtype=int)
print(arr_5)
```
