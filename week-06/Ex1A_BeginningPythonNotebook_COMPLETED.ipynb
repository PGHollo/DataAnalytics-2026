```
sales = [120, 300, 250, 175]
total = sum(sales)
average = total / len(sales)
print(f'Total sales: ${total:,.2f}')
print(f'Average sale: ${average:,.2f}')
```

```
def revenue_report(amounts):
    total = sum(amounts)
    average = total / len(amounts)
    return f'Total: ${total:,.2f} | Average: ${average:,.2f}'

print(revenue_report(sales))
```
