# Pi Birthday Bonus
This checks whether a birthday appears in the first million digits of pi.

```
try:
    pi = open('pi_million_digits_bonus.txt', 'r')
    print(pi.readline())
    pi.close()
except FileNotFoundError:
    print('pi_million_digits_bonus.txt is missing. Download it into this folder before running the full lab.')
```

```
try:
    pi = open('pi_million_digits_bonus.txt', 'r')
    pi_lines = pi.readlines()
    pi.close()
except FileNotFoundError:
    pi_lines = ['3.14159265358979323846264338327950288419716939937510']
```

```
def pi_bday_check(bday):
    for line in pi_lines:
        if bday in line:
            print('Your birthday is in the first million digits of pi')
            break

pi_bday_check('070281')
```

```
def pi_bday_check2(bday):
    birthday_found = 0

    for line in pi_lines:
        if bday in line:
            print('Your birthday is in the first million digits of pi')
            birthday_found += 1
            break

    if birthday_found != 1:
        print('Sorry, your birthday was not found in the first million digits of pi')

pi_bday_check2('070281')
pi_bday_check2('112952')
```

```
pi_string = ''

for line in pi_lines:
    pi_string += line.strip()

print(pi_string[0:300])
```

```
try:
    birthday_position = pi_string.index('070281') - 1
    # The -1 adjusts for Python starting at index 0 and for the 3 before the decimal.
    print(birthday_position)
except ValueError:
    print('Birthday not found in the loaded pi text.')
```

```
def pi_bday_check3(bday):
    birthday_found = 0
    bday_position = ''

    for line in pi_lines:
        if bday in line:
            print('Your birthday is in the first million digits of pi')
            birthday_found += 1
            break

    if birthday_found != 1:
        print('Sorry, your birthday was not found in the first million digits of pi')
    else:
        bday_position = pi_string.index(bday) - 1
        print(f'Your birthday begins at decimal place {bday_position}')

pi_bday_check3('070281')
```
