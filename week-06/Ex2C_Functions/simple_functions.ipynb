Working with defining custom functions.

```python
def favorite_things(name, movie, music):
    line1 = f'Name: {name}'
    line2 = f'Favorite movie: {movie}'
    line3 = f'Favorite musician/band: {music}'
    return line1 + '\n' + line2 + '\n' + line3

print(favorite_things('Patrick Green-Holloway', 'Black Panther', 'Kanye West'))
```

```python
def why_im_here(purpose):
    return purpose

message = 'I joined Year Up United to learn versatile coding skills and build a better career.'
print(why_im_here(message))
```

```python
def favorite_place(place, why):
    return f'One of my favorite places is {place} because {why}.'
```

```python
print(favorite_place('Baltimore Inner Harbor', 'it has good views, food, and a chill city vibe'))
```

```python
print(favorite_place('Crafty Crab', 'seafood is one of my favorite meals to share with people'))
```
