# Exercise 2.C Lab 3
This lab creates payroll tax functions and tests them with sample gross pay amounts.

```
def get_soc_sec_tax(gross_pay):
    tax = gross_pay * 0.062
    return tax
```

```
def get_medicare_tax(gross_pay):
    tax = gross_pay * 0.0145
    return tax
```

```
def get_federal_tax(gross_pay, withholding_code):
    if withholding_code == 0:
        rate = 0.23
    elif withholding_code == 1:
        rate = 0.21
    elif withholding_code == 2:
        rate = 0.195
    elif withholding_code == 3:
        rate = 0.185
    else:
        extra_withholding = withholding_code - 4
        rate = 0.18 - (extra_withholding * 0.005)

    tax = gross_pay * rate
    return tax
```

```
gross_pay = 750
withholding_code = 0
print(f'Social Security: ${get_soc_sec_tax(gross_pay):.2f}')
print(f'Medicare: ${get_medicare_tax(gross_pay):.2f}')
print(f'Federal: ${get_federal_tax(gross_pay, withholding_code):.2f}')
```
