## Looping Through Arrays

```python
import numpy as np

arrA = np.array([
    [11, 12, 13],
    [14, 15, 16]
])

print(arrA)
```

```python
for x in arrA:
    print(x * 2)
```

```python
for y in arrA:
    print(y * 2)
```

## They gave the same result?

```python
for x in arrA:
    for y in x:
        loop_result = y * 2
        print(loop_result)

# It printed the results vertically this time because we are printing each element of the array separately, rather than printing the entire row at once. When we print the entire row, it is displayed horizontally as a single line. However, when we print each element individually, they are printed on separate lines, resulting in a vertical display of the results.
```

## Joining and Stacking Arrays

```python
joined_array_1 = np.concatenate((arrA, arrA), axis=0)
print(joined_array_1)
```

```python
joined_array_2 = np.concatenate((arrA, arrA), axis=1)
print(joined_array_2)
```

```python
stacked_array_1 = np.stack((joined_array_1, joined_array_1), axis=0)
print(stacked_array_1)
```

```python
stacked_array_2 = np.stack((joined_array_2, joined_array_2), axis=1)
print(stacked_array_2)
```

```python
another_joined_array_1 = np.concatenate((joined_array_1, arrA), axis=0)
print(another_joined_array_1)
```

```python
another_joined_array_2 = np.concatenate((joined_array_1, arrA), axis=1)
print(another_joined_array_2)

#All the input array dimensions must match except for the concatenation axis. The joined array will have the same number of dimensions as the input arrays. The shape of the joined array will be the same as the input arrays, except along the concatenation axis, where it will be the sum of the sizes of the input arrays along that axis.
```

```python
h_stack = np.hstack((arrA, arrA))  # hstack stacks arrays horizontally, side by side.
v_stack = np.vstack((arrA, arrA))  # vstack stacks arrays vertically, one row on top of another.
d_stack = np.dstack((arrA, arrA))  # dstack stacks arrays by depth, creating a 3D-style array.

print("Horizontal Stack:")
print(h_stack)

print("\nVertical Stack:")
print(v_stack)

print("\nDepth Stack:")
print(d_stack)
```

## Splitting Arrays

```python
splitsky = np.split(joined_array_1, 2)
print(splitsky)

#This split the array into 2 1 dimensional arrays.
```

```python
splitereno = np.split(joined_array_1, 2)
print(splitereno)

# This segments the array into 2 2-dimensional arrays. Each resulting array will contain half of the rows of the original array.
```

```python
splitereno = np.split(joined_array_1, 3)
print(splitereno)

# This gives a value error because the array cannot be split into 3 equal parts along the specified axis. The number of splits must be a divisor of the size of the array along that axis. In this case, since joined_array_1 has a shape that does not allow for an equal split into 3 parts, it raises an error.
```

## Searching and Replacing Array Values.

```python
result_1 = np.where(arrA % 2 == 0, arrA, "odd")
print(result_1)
```
