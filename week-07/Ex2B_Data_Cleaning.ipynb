## Import pandas

```python
import pandas as pd

# Bringing pandas in so I can work with tables/DataFrames.
```

## Create / load the test data

```python
# This is the same data from the last exercise.
# I am saving it as a CSV first so this notebook can run on its own.

test_data = {
    "ID": [1, 2, 3, 4, 5, 6, 7, 8],
    "Name": ["Alicia", "Bob", "Charlie", "Dominic", "Eve", "Frank", "Grace", "Heidi"],
    "Age": [25, None, 22, 19, 31, 35, None, 18],
    "City": ["NY", "LA", "Chicago", "Houston", "Phoenix", "Boston", "Austin", "San Diego"],
    "Score": [85, 92, -1, 65, 78, None, 55, 90],
}

pd.DataFrame(test_data).to_csv("test_data.csv", index=False)

# Now I am reading the CSV file into a DataFrame named df.
df = pd.read_csv("test_data.csv")

# Checking the full DataFrame so I know the data loaded correctly.
df
```

## Column statistics

```python
# describe() gives quick stats like count, mean, min, and max.
# This helps me spot missing values or weird numbers before cleaning.
df.describe()
```

## Filtering a DataFrame

```python
# This shows only people older than 18.
# I used > 18 because the task says greater than 18, not equal to 18.
df[df["Age"] > 18]
```

```python
# This keeps scores that are between 50 and 100.
# It helps remove scores that are too low, too high, or just not valid.
df[(df["Score"] > 50) & (df["Score"] < 100)]
```

## Modifying a DataFrame

```python
# I am making a cleaned copy of df.
# The original df stays the same, and df_cleaned is the one I clean up.
# This drops rows where Score is less than 0 because negative scores do not make sense here.

df_cleaned = df.drop(df[df["Score"] < 0].index)

# Reviewing the cleaned DataFrame after removing the invalid score row.
df_cleaned
```

```python
# Dropping the City column because the task says to remove it.
# axis=1 means I am dropping a column, not a row.

df_cleaned = df_cleaned.drop("City", axis=1)

# Checking that City is gone now.
df_cleaned
```

```python
# Filling in missing Age values with the average Age.
# This is useful because it keeps the rows instead of deleting them.

mean_age = df_cleaned["Age"].mean()
df_cleaned["Age"] = df_cleaned["Age"].fillna(mean_age)

# Checking the DataFrame to make sure the missing Ages are filled in.
df_cleaned
```

```python
# Replacing missing Score values with the average Score.
# This is another way to handle missing data without losing the whole row.

mean_score = df_cleaned["Score"].mean()
df_cleaned["Score"] = df_cleaned["Score"].replace(pd.NA, mean_score)

# Since pandas may read missing numbers as NaN, this makes sure all missing scores are filled.
df_cleaned["Score"] = df_cleaned["Score"].fillna(mean_score)

# Checking the DataFrame after replacing the missing Score.
df_cleaned
```

```python
# dropna() removes rows that still have missing values.
# At this point, there should not be any left, but I am checking/cleaning anyway.

df_cleaned = df_cleaned.dropna()

# Reviewing the DataFrame after removing any leftover NaN values.
df_cleaned
```

```python
# Setting ID as the index because ID works like a unique label for each row.
# Then I remove the ID column from the regular columns.

df_cleaned = df_cleaned.set_index("ID")

# Final review of the cleaned DataFrame.
df_cleaned
```

## Quick final notes

The main cleaning steps were: remove bad scores, remove a column we do not need, fill in missing numbers with averages, remove leftover missing values, and use ID as the row label.
