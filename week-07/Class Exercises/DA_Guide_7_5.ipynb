```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
```

```python
# ── Sample DataFrame: 5 regions × 6 months of sales data ─────────────
df_sales = pd.DataFrame({
 'region' : ['North', 'South', 'East', 'West', 'Central'],
 'jan' : [12400, 9800, 15200, 8700, 11300],
 'feb' : [13100, 10500, 14800, 9200, 12100],
 'mar' : [14500, 11200, 16300, 10100, 13400],
 'apr' : [13800, 10900, 15700, 9600, 12800],
 'may' : [15200, 11800, 17100, 10400, 14200],
 'jun' : [16000, 12400, 18500, 11000, 15600]
})
months_cols = ['jan', 'feb', 'mar', 'apr', 'may', 'jun']
df_sales['total_sales'] = df_sales[months_cols].sum(axis=1)
print(df_sales)
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
north = df_sales.loc[df_sales['region'] == 'North', 
months_cols].values[0].tolist()
# Step 1 — Create canvas
fig, ax = plt.subplots(figsize=(10, 5))
# Step 2 — Draw the line
ax.plot(months, north,
 color='#2196F3', # line colour
 linewidth=2.5, # line thickness
 marker='o', # circle at each data point
 markersize=8,
 markerfacecolor='white',# hollow marker fill
 markeredgewidth=2,
 linestyle='-', # solid line (use '--' for dashed)
 label='North'
)
# Step 3 — Labels and title
ax.set_title('North Region — Monthly Sales Trend H1 2024',
 fontsize=14, fontweight='bold', pad=12)
ax.set_xlabel('Month', fontsize=12)
ax.set_ylabel('Sales (£)', fontsize=12)
# Step 4 — Formatting extras
ax.yaxis.set_major_formatter(
 mticker.FuncFormatter(lambda x, _: f'£{x:,.0f}'))
ax.grid(axis='y', alpha=0.3, linestyle='--')
ax.set_ylim(0, 20000)
# Add a mean reference line
mean_val = sum(north) / len(north)
ax.axhline(mean_val, color='red', ls='--', alpha=0.6, label=f'Mean £{mean_val:,.0f}')
ax.legend(fontsize=10)
# Step 5 — Render
plt.tight_layout()
plt.show()
```
