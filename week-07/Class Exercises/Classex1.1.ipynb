```python
# Simple Line Graph
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create line plot
plt.plot(x, y, scalex=True, scaley=True)

# Add labels and title
plt.title("Simple Line Graph")
plt.xlabel("X values")
plt.ylabel("Y values")

# Display grid (optional but helpful)
plt.grid(True)

# Show the plot
plt.show()
```

```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt

sales = pd.DataFrame({
    "month": [1, 2, 3, 4, 5, 6],
    "store_A": [45, 34, 62, 55, 69, 72],
    "store_B": [30, 40, 50, 45, 60, 65]
})

# Set figure size
plt.figure(figsize=(4, 2))

# Plot multiple lines from DataFrame
plt.plot(sales["month"], sales["store_A"], marker="o", lw=3, label="Store A")
plt.plot(sales["month"], sales["store_B"], marker="x", lw=3, linestyle="--", label="Store B")

# Titles and labels
plt.title("Monthly Sales Comparison")
plt.xlabel("Month")
plt.ylabel("Quantity Sold")

# Grid + Legend
plt.grid(True)
plt.legend()

out_path = Path(r"G:/Yu_Class 7/PythonFiles2026/Week7/line_graph4.png")
out_path.parent.mkdir(parents=True, exist_ok=True)
plt.savefig(out_path)

# Show plot
plt.show()
```
