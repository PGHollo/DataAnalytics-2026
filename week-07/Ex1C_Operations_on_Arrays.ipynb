## Sample Arrays

```python
# I am making the same random arrays from Exercise 1.B so I can use them in this lab.
# The seed makes the random numbers stay the same each time I run the notebook.
np.random.seed(7)

r_1 = np.random.randint(1, 21, size=2)
r_2 = np.random.randint(1, 21, size=(1, 2))
r_3 = np.random.random(size=(2, 1))
r_4 = np.random.randint(20, 101, size=(4, 5))
r_5 = np.random.random(size=(4, 5))

print("r_1:")
print(r_1)

print("\nr_2:")
print(r_2)

print("\nr_3:")
print(r_3)

print("\nr_4:")
print(r_4)

print("\nr_5:")
print(r_5)
```

## Permutations

```python
# This shows the original r_4 before I shuffle anything.
print("Original r_4:")
print(r_4)
```

```python
# np.random.shuffle changes r_4 directly.
# It shuffles the rows around, but the numbers inside each row stay together.
np.random.shuffle(r_4)

print("r_4 after shuffle:")
print(r_4)
```

After using `shuffle`, the rows in `r_4` moved around. The important thing I noticed is that `shuffle` changed the original array instead of making a separate new copy.

```python
# np.random.permutation makes a shuffled copy of r_4.
# This means new_shuffle is changed, but r_4 stays the same.
new_shuffle = np.random.permutation(r_4)

print("new_shuffle:")
print(new_shuffle)

print("\nr_4 for comparison:")
print(r_4)
```

I would use `shuffle` when I am okay with changing the original array. I would use `permutation` when I want a shuffled version but still want to keep the original array the same.

## Mathematical Operations

```python
# I am checking the starting values and shapes first.
# Shape helps me know if the arrays can be added together.
print("r_1:")
print(r_1)
print("r_1 shape:", r_1.shape)

print("\nr_2:")
print(r_2)
print("r_2 shape:", r_2.shape)

result = np.add(r_1, r_2)
print("\nr_1 plus r_2:")
print(result)
print("result shape:", result.shape)
```

```python
# I am adding r_1 and r_3.
# NumPy broadcasts them because their shapes can stretch together.
print("r_1:")
print(r_1)
print("r_1 shape:", r_1.shape)

print("\nr_3:")
print(r_3)
print("r_3 shape:", r_3.shape)

result = np.add(r_1, r_3)
print("\nr_1 plus r_3:")
print(result)
print("result shape:", result.shape)
```

```python
# I am adding r_2 and r_4.
# r_2 has 2 columns, but r_4 has 5 columns, so this should cause a shape problem.
print("r_2:")
print(r_2)
print("r_2 shape:", r_2.shape)

print("\nr_4:")
print(r_4)
print("r_4 shape:", r_4.shape)

try:
    result = np.add(r_2, r_4)
    print("\nr_2 plus r_4:")
    print(result)
    print("result shape:", result.shape)
except ValueError as error:
    print("Error message:", error)
```

```python
# This shows what happens when r_1 and r_4 are added.
# r_1 has 2 values, but r_4 has 5 columns, so they do not line up.
try:
    print(np.add(r_1, r_4))
except ValueError as error:
    print("Error message for r_1 + r_4:", error)

# This also does not work because r_2 has 2 columns and r_4 has 5 columns.
try:
    print(np.add(r_2, r_4))
except ValueError as error:
    print("Error message for r_2 + r_4:", error)

# This works because r_3 has 2 rows and 1 column, so it can stretch across r_4's columns.
try:
    print("\nr_3 + r_4 works:")
    print(np.add(r_3, r_4))
except ValueError as error:
    print("Error message for r_3 + r_4:", error)

# My finding: the arrays need shapes that can match or stretch using broadcasting.
```

```python
# Adding np.array([1]) works because one number can be added to every value in r_4.
print(np.add(np.array([1]), r_4))
```

```python
# This also works because [[1]] is still basically one value that can stretch across r_4.
print(np.add(np.array([[1]]), r_4))
```

```python
# This does not work because this array has 4 values, but r_4 has 5 columns.
try:
    print(np.add(np.array([1, 1, 1, 1]), r_4))
except ValueError as error:
    print("Error message:", error)
```

```python
# This does not work because the shape is 2 rows by 1 column.
# r_4 has 4 rows by 5 columns, so the row count does not match.
try:
    print(np.add(np.array([[1], [1]]), r_4))
except ValueError as error:
    print("Error message:", error)

# My finding: NumPy can stretch size 1, but it cannot stretch size 2 into size 4.
```

Broadcasting is how NumPy adds arrays that are not exactly the same shape. A small array can stretch across a bigger array when the shapes make sense. With the `add` function, NumPy adds matching positions together. The resulting shape usually comes from the larger array after the smaller one gets stretched.

## Cumulative Operations

```python
# This is the array I will use for sums and cumulative sums.
print(r_4)
```

```python
# This adds every single value in r_4 together.
print("Total sum:", np.sum(r_4))
```

```python
# Axis 0 adds down the columns.
print("Sum along axis 0:")
print(np.sum(r_4, axis=0))
```

```python
# Axis 1 adds across each row.
print("Sum along axis 1:")
print(np.sum(r_4, axis=1))
```

```python
# cumsum gives a running total through the whole array.
print("Cumulative sum:")
print(np.cumsum(r_4))
```

```python
# Axis 0 cumulative sum gives a running total going down each column.
print("Cumulative sum along axis 0:")
print(np.cumsum(r_4, axis=0))
```

## Grocery Problem

```python
# The order is: vegetables, fruits, boxes of snacks, beverages.
# bag1 has 2 veggies, 3 fruits, 4 snacks, and 0 drinks.
# bag2 has 4 veggies, 2 fruits, 1 snack, and 1 drink.
bag1 = np.array([2, 3, 4, 0])
bag2 = np.array([4, 2, 1, 1])

print("bag1:", bag1)
print("bag2:", bag2)
```

```python
# I use sum to count all the items in each bag.
bag1_total = np.sum(bag1)
bag2_total = np.sum(bag2)

print(f"The first bag contains {bag1_total} items, and the second bag contains {bag2_total} items.")
```

```python
# Adding the bags together gives totals by category.
category_totals = np.add(bag1, bag2)

vegetables, fruits, snacks, beverages = category_totals

print(
    f"There are {vegetables} vegetables, {fruits} fruits, "
    f"{snacks} boxes of snacks, and {beverages} beverages in the two grocery bags."
)
```

## Teamwork

```python
# I am using 0 for missed days so each person still has Monday through Friday.
# The order is: Monday, Tuesday, Wednesday, Thursday, Friday.
ali = np.array([5, 4, 6, 4, 2])
ben = np.array([0, 5, 5, 3, 4])
carla = np.array([3, 6, 0, 3, 4])

print("Ali:", ali)
print("Ben:", ben)
print("Carla:", carla)
```

```python
# vstack puts each person's numbers into one bigger array.
# Each row is one person, and each column is one day.
all_knocks = np.vstack([ali, ben, carla])

print("all_knocks:")
print(all_knocks)
```

```python
# Axis 1 gives each person's weekly total.
# The regular sum gives the whole team's total.
person_totals = np.sum(all_knocks, axis=1)
group_total = np.sum(all_knocks)

print("Ali knocked on", person_totals[0], "doors this week.")
print("Ben knocked on", person_totals[1], "doors this week.")
print("Carla knocked on", person_totals[2], "doors this week.")
print("The whole group knocked on", group_total, "doors total.")
```

```python
# First I add by day using axis 0.
# Then I use cumsum to get the running total by day.
# Since 1 out of every 2 door-knocks may turn into a vote, I multiply by 0.5.
daily_totals = np.sum(all_knocks, axis=0)
rolling_knocks = np.cumsum(daily_totals)
estimated_votes = rolling_knocks * 0.5

print("Daily door-knock totals:", daily_totals)
print("Rolling door-knock totals:", rolling_knocks)
print("Rolling estimated votes:", estimated_votes)
```
