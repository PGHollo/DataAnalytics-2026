## Import NumPy

## Generating Random Numbers

### r_1 — Random integers, 1-D array
Create 2 random integers from 1 to 20.

```python
r_1 = np.random.randint(1, 21, size=2)
print(r_1)
```

### r_2 — Random integers, 1 row by 2 columns
Create random integers from 1 to 20 in a 2-D shape.

```python
r_2 = np.random.randint(1, 21, size=(1, 2))
print(r_2)
```

### r_3 — Random floats, 2 rows by 1 column
Create random decimal values between 0 and 1.

```python
r_3 = np.random.random(size=(2, 1))
print(r_3)
```

### r_4 — Random integers, 4 rows by 5 columns
Create random integers from 20 to 100.

### r_5 — Random floats, 4 rows by 5 columns
Create random decimal values between 0 and 1.

## Generating Random Choices from an Array

### students array
Create a 1-D array with 12 classmate names.

```python
students = np.array([
    "Alice", "Brian", "Carlos", "Destiny",
    "Elijah", "Fatima", "Grace", "Hector",
    "Imani", "Jordan", "Kayla", "Luis"
])

print(students)
```

### call_on — choose one random student
Randomly select one person from the students array.

```python
call_on = np.random.choice(students)
print(call_on)
```

### breakout_groups — create 3 groups with 4 students each
Randomly place the 12 students into a 3 by 4 array.

```python
breakout_groups = np.random.choice(students, size=(3, 4), replace=False)
print(breakout_groups)

# Yes, this is an effective way to assign breakout groups because replace=False
# means each student can only be selected once, so no one appears in two groups.
```

### call_on_weighted — choose 5 students with different probabilities
Use probability values to control how likely each student is to be selected.

```python
probabilities = [
    0.30, 0.15, 0.10, 0.10,
    0.08, 0.07, 0.06, 0.05,
    0.04, 0.03, 0.02, 0.00
]

call_on_weighted = np.random.choice(
    students,
    size=5,
    replace=True,
    p=probabilities
)

print(call_on_weighted)

# Probability explanation:
# Alice has a 30% chance of being selected, so Alice is the most likely.
# Luis has a 0% chance of being selected, so Luis will not be called on.
# The other students have smaller probability values that add up to 70%.
```

## Generating a Normal Distribution

### bell_curve — normal distribution
Create a random normal distribution where the mean is 10, the standard deviation is 5, and the shape is 4 rows by 5 columns.

```python
bell_curve = np.random.normal(loc=10, scale=5, size=(4, 5))
print(bell_curve)
```
