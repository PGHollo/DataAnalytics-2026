# Exercise 3.A Getting familiar with Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt

# Bringing in NumPy and Matplotlib so I can make charts in this notebook.
```

# Creating Basic Plots

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 7, 11]

# These are the simple x and y values I am using for the graph.
print("x values:", x)
print("y values:", y)
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y)
plt.show()

# This makes a basic line chart using my x and y lists.
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y, marker='o')
plt.show()

# The circle markers make each data point easier to see on the line.
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y, marker='o')
plt.title('Example Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()

# The title and labels help explain what the chart is showing.
```

# Styling Plots

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y, linestyle='--', color='red')
plt.title('Dashed Red Line Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()

# I changed the line to dashed and red so the graph has a different style.
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y, marker='s', markersize=10, markerfacecolor='green')
plt.title('Square Marker Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()

# The square markers are bigger and green, so the points stand out more.
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y, marker='^', markersize=9, color='purple')
plt.title('Triangle Marker Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()

# This is my first extra style using triangle markers and a purple line.
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y, marker='D', markersize=8, color='orange')
plt.title('Diamond Marker Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()

# This is my second extra style using diamond markers and an orange line.
```
