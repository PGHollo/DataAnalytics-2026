This notebook uses NumPy to analyze weekly revenue data for five regional offices.
It calculates statistics, applies a 5% bonus, filters high-performing regions, and reshapes the data for 2-D analysis.

```python
import numpy as np

revenue = np.array([52_400.0, 38_750.0, 61_200.0, 47_900.0, 55_650.0])

print(f"Array: {revenue}")
print(f"Dtype: {revenue.dtype}")
print(f"Shape: {revenue.shape}")
```

```python
print("Revenue Statistics")
print(f"Total Revenue: ${np.sum(revenue):,.2f}")
print(f"Mean Revenue: ${np.mean(revenue):,.2f}")
print(f"Median Revenue: ${np.median(revenue):,.2f}")
print(f"Standard Deviation: ${np.std(revenue):,.2f}")
print(f"Minimum Revenue: ${np.min(revenue):,.2f}")
print(f"Maximum Revenue: ${np.max(revenue):,.2f}")
```

```python
revenue_bonus = revenue * 1.05
bonus_amount = revenue_bonus - revenue

print(f"Original Revenue: {revenue}")
print(f"Revenue With 5% Bonus: {revenue_bonus}")
print(f"Bonus Amount Per Region: {bonus_amount}")
```

```python
above_avg = revenue > np.mean(revenue)

print(f"Above Average Mask: {above_avg}")
print(f"Above Average Revenue Values: {revenue[above_avg]}")
print(f"Number of Regions Above Average: {np.sum(above_avg)}")
```

```python
revenue_column = revenue.reshape(5, 1)

print("Revenue Column Array:")
print(revenue_column)

q_data = np.vstack([revenue, revenue])

print("\n2-D Revenue Data:")
print(q_data)

row_totals = np.sum(q_data, axis=1)
column_means = np.mean(q_data, axis=0)

print(f"\nRow Totals: {row_totals}")
print(f"Column Region Means: {column_means}")
```

```python
import numpy as np

revenue = np.array([52_400.0, 38_750.0, 61_200.0, 47_900.0, 55_650.0])

print(f"Array: {revenue}")
print(f"Dtype: {revenue.dtype}")
print(f"Shape: {revenue.shape}")

print("\nRevenue Statistics")
print(f"Total Revenue: ${np.sum(revenue):,.2f}")
print(f"Mean Revenue: ${np.mean(revenue):,.2f}")
print(f"Median Revenue: ${np.median(revenue):,.2f}")
print(f"Standard Deviation: ${np.std(revenue):,.2f}")
print(f"Minimum Revenue: ${np.min(revenue):,.2f}")
print(f"Maximum Revenue: ${np.max(revenue):,.2f}")

revenue_bonus = revenue * 1.05
bonus_amount = revenue_bonus - revenue

print(f"\nOriginal Revenue: {revenue}")
print(f"Revenue With 5% Bonus: {revenue_bonus}")
print(f"Bonus Amount Per Region: {bonus_amount}")

above_avg = revenue > np.mean(revenue)

print(f"\nAbove Average Mask: {above_avg}")
print(f"Above Average Revenue Values: {revenue[above_avg]}")
print(f"Number of Regions Above Average: {np.sum(above_avg)}")

revenue_column = revenue.reshape(5, 1)

print("\nRevenue Column Array:")
print(revenue_column)

q_data = np.vstack([revenue, revenue])

print("\n2-D Revenue Data:")
print(q_data)

row_totals = np.sum(q_data, axis=1)
column_means = np.mean(q_data, axis=0)

print(f"\nRow Totals: {row_totals}")
print(f"Column Region Means: {column_means}")
```
