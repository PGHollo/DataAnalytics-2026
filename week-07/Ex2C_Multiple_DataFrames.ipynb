# Exercise 2.C Working with multiple DataFrames

```python
# Importing the tools I need for this exercise
import numpy as np
import pandas as pd
```

## Basic Aggregations

```python
# This is the starting sales data for the exercise.
# Each row shows a product, region, units sold, and price.
sales_data = {
    'Product': ['A', 'B', 'C', 'A', 'B', 'C', 'A', 'B', 'C'],
    'Region': ['North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 'North'],
    'Units_Sold': [10, 15, 20, 5, 10, 15, 10, 5, 15],
    'Price': [100, 200, 150, 100, 200, 150, 100, 200, 150]
}

# Turning the dictionary into a DataFrame so it looks like a small table.
df_sales = pd.DataFrame(sales_data)

df_sales
```

```python
# Revenue is units sold times price.
# This adds a new column so I can see how much money each row made.
df_sales['Revenue'] = df_sales['Units_Sold'] * df_sales['Price']

df_sales
```

```python
# This gives the total units sold and total revenue for the whole DataFrame.
# Basically, it adds everything together across all products and regions.
total_units_sold = df_sales['Units_Sold'].sum()
total_revenue = df_sales['Revenue'].sum()

print(f"Total Units Sold: {total_units_sold}")
print(f"Total Revenue: ${total_revenue:,.2f}")
```

```python
# This finds the average units sold and average revenue.
# I am using mean() because it gives the average value.
average_units_sold = df_sales['Units_Sold'].mean()
average_revenue = df_sales['Revenue'].mean()

print(f"Average Units Sold: {average_units_sold:.2f}")
print(f"Average Revenue: ${average_revenue:,.2f}")
```

## Grouping Data

```python
# This groups the data by Product.
# It shows the total units sold and revenue for each product.
product_totals = df_sales.groupby('Product')[['Units_Sold', 'Revenue']].sum()

product_totals
```

```python
# This groups the data by Region.
# It shows the average units sold and average revenue for each region.
region_averages = df_sales.groupby('Region')[['Units_Sold', 'Revenue']].mean()

region_averages
```

```python
# This groups by Product and Region together.
# It helps me see the totals for each product in each region.
product_region_totals = df_sales.groupby(['Product', 'Region'])[['Units_Sold', 'Revenue']].sum()

product_region_totals
```

## Sorting and Ranking Data

```python
# This sorts the rows from the lowest Units_Sold to the highest.
# Ascending means the smaller numbers show up first.
sorted_by_units = df_sales.sort_values(by='Units_Sold', ascending=True)

sorted_by_units
```

```python
# This sorts the rows by Revenue from highest to lowest.
# I use descending order because I want to see the biggest revenue first.
sorted_by_revenue = df_sales.sort_values(by='Revenue', ascending=False)

sorted_by_revenue
```

```python
# This ranks each row based on Revenue.
# Rank 1 means that row made the most revenue.
df_sales['Revenue_Rank'] = df_sales['Revenue'].rank(ascending=False)

df_sales
```

```python
# This shows the DataFrame with the new Revenue_Rank column.
# Now I can quickly spot which rows performed the best.
df_sales
```

## Merging DataFrames

```python
# This is a second DataFrame with product category information.
# It connects each product letter to a category name.
product_data = {
    'Product': ['A', 'B', 'C'],
    'Category': ['Electronics', 'Furniture', 'Appliances']
}

df_products = pd.DataFrame(product_data)

df_products
```

```python
# This merges the sales DataFrame with the products DataFrame.
# They match together using the Product column.
df_merged = pd.merge(df_sales, df_products, on='Product')

df_merged
```

```python
# Calling df_merged again so I can review the final merged table.
# This now includes sales info plus the product category.
df_merged
```

```python
# This groups the merged data by Category.
# It shows how much total revenue each category made.
category_revenue = df_merged.groupby('Category')['Revenue'].sum()

category_revenue
```
