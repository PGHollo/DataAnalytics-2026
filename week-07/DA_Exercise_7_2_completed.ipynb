This notebook uses Pandas to create, inspect, filter, and summarize employee sales data for RetailHub Inc. It practices DataFrame creation, Boolean filtering, calculated columns, lambda functions, and GroupBy aggregation.

## 1. Import Pandas and Create the DataFrame
We import Pandas using the standard `pd` alias and build a DataFrame from a Python dictionary of employee sales records.

```python
# import Pandas and create the DataFrame
import pandas as pd

# Employee sales data: 8 records and 4 columns
data = {
    'name': ['Alice', 'Ben', 'Carmen', 'Dev', 'Eve', 'Frank', 'Grace', 'Hiro'],
    'city': ['London', 'Manchester', 'London', 'Leeds',
             'Birmingham', 'Manchester', 'Leeds', 'London'],
    'sales': [4200.0, 3100.0, 5050.0, 2800.0,
              1950.0, 3750.0, 4600.0, 3300.0],
    'months': [12, 6, 24, 3, 18, 9, 15, 7]
}

df = pd.DataFrame(data)

print('── Employee DataFrame ──────────────────────')
print(df)
print(f"\nShape: {df.shape}")
print("\nData types:")
print(df.dtypes)
print("\ndf.info():")
df.info()
```

## 2. Inspect and Describe the Data
We preview the first and last rows, then use summary statistics to understand the numeric columns.

```python
# inspect and describe the data
print('First 3 rows:')
print(df.head(3))

print('Last 3 rows:')
print(df.tail(3))

print('Summary statistics:')
print(df.describe())

print(f'Mean sales: £{df["sales"].mean():,.2f}')
print(f'Max sales: £{df["sales"].max():,.2f}')
```

## 3. Select, Filter, and Sort
We select specific columns, filter employees above a sales amount, sort the data, and display the top seller.

```python
# select, filter, and sort
name_sales = df[['name', 'sales']]
print('Name and Sales columns:')
print(name_sales)

high_sales = df[df['sales'] > 3500.0]
print('Employees with sales above £3,500:')
print(high_sales)

df_sorted = df.sort_values('sales', ascending=False).reset_index(drop=True)
print('Full DataFrame sorted by sales, highest first:')
print(df_sorted)

print('Top-selling employee full record:')
print(df_sorted.iloc[0])
```

## 4. Add a Calculated Column and Conditional Rating
We calculate each employee's 8% commission and add a performance rating based on their sales amount.

```python
# add commission and rating columns

df['commission'] = df['sales'] * 0.08

df['rating'] = df['sales'].apply(
    lambda sale: 'Star' if sale >= 4000 else ('Good' if sale >= 2500 else 'Developing')
)

print('Updated DataFrame with commission and rating:')
print(df)

print('── Commission per Employee ──────────────')
for _, row in df.iterrows():
    print(f'{row["name"]:<8} | £{row["commission"]:>8,.2f} | {row["rating"]}')
```

## 5. GroupBy — Aggregate Statistics by City
We group the employees by city to compare total sales, average sales, and employee counts by location.

```python
# GroupBy aggregate statistics by city

city_totals = df.groupby('city')['sales'].sum()
print('── Total Sales by City ──────────────────')
print(city_totals)

city_means = df.groupby('city')['sales'].mean()
print('── Mean Sales by City ───────────────────')
print(city_means)

city_summary = df.groupby('city').agg({'sales': ['sum', 'mean', 'count']})
print('── Full City Summary ────────────────────')
print(city_summary)

top_city = city_totals.idxmax()
print(f'Top city: {top_city} Total: £{city_totals[top_city]:,.2f}')
```

## Summary
London is the top-performing city with total sales of **£12,550.00** from 3 employees. The strongest individual performer is **Carmen**, who made **£5,050.00** in sales and earned a **Star** rating. Based on the GroupBy results, RetailHub should continue investing in London while looking for ways to improve Birmingham because it has the lowest total sales.

## Extension Challenge 1 — Sales Bands
This extra cell groups sales into three equal-width bands and counts how many employees fall into each band.

```python
sales_bands = df['sales'].value_counts(bins=3)

print('Sales grouped into 3 bands:')
print(sales_bands)
```

## Extension Challenge 2 — Pivot Table
This extra cell creates a pivot table showing average sales by city and compares it to the GroupBy mean result.

```python
pivot_sales = pd.pivot_table(df, values='sales', index='city', aggfunc='mean')

print('Pivot table — average sales by city:')
print(pivot_sales)

print('\nGroupBy mean — average sales by city:')
print(city_means)
```

## Extension Challenge 3 — Rank by Tenure
This extra cell sorts employees by months with the company and adds a rank column.

```python
tenure_rank = df.sort_values('months').reset_index(drop=True)
tenure_rank['rank'] = tenure_rank.index + 1

print('Employees ranked by tenure:')
print(tenure_rank)
```

## Extension Challenge 4 — Star Performers
This extra cell filters only Star performers and calculates their combined commission.

```python
star_performers = df[df['rating'] == 'Star'][['name', 'city', 'sales', 'commission']]
combined_star_commission = star_performers['commission'].sum()

print('Star performers:')
print(star_performers)
print(f'Combined Star performer commission: £{combined_star_commission:,.2f}')
```

## Extension Challenge 5 — Missing Data Practice
This extra cell simulates a missing sales value, detects it, fills it with the mean, and prints the cleaned DataFrame.

```python
missing_df = df.copy()
missing_df.loc[2, 'sales'] = None

print('Missing values before cleaning:')
print(missing_df.isnull().sum())

missing_df['sales'] = missing_df['sales'].fillna(missing_df['sales'].mean())

print('Missing values after cleaning:')
print(missing_df.isnull().sum())

print('Cleaned DataFrame:')
print(missing_df)
```
