# Exercise 3.B More plotting practice

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Advanced Plotting Techniques

```python
products = ['Product A', 'Product B', 'Product C', 'Product D']
sales = [23, 17, 35, 29]

# This makes a basic bar chart so I can compare each product's sales side by side.
plt.bar(products, sales, color='skyblue', alpha=0.5)

# These labels help the chart make sense instead of just showing random bars.
plt.title('Sales by Product')
plt.xlabel('Products')
plt.ylabel('Sales')

# This displays the chart after everything is set up.
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)

# This creates random sizes so each dot is a little different on the scatter plot.
sizes = np.random.rand(50) * 100

# This scatter plot helps show if there is any pattern between x and y values.
plt.scatter(x, y, s=sizes, c=y, cmap='viridis')

# These labels make the scatter plot easier to read.
plt.title('Random Scatter Plot')
plt.xlabel('X Values')
plt.ylabel('Y Values')

# This color bar helps explain what the colors mean on the plot.
plt.colorbar(label='Y Value Color Scale')

# This displays the finished scatter plot.
plt.show()
```

## Subplots

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

# This creates two chart spaces in one figure, so I can show two plots together.
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# First subplot: this line chart shows a sine wave.
axes[0].plot(x, y)
axes[0].set_title('Sine Wave')
axes[0].set_xlabel('X Values')
axes[0].set_ylabel('sin(x)')

# Second subplot: this histogram shows how random numbers are spread out.
random_numbers = np.random.randn(1000)
axes[1].hist(random_numbers)
axes[1].set_title('Histogram of Random Numbers')
axes[1].set_xlabel('Value')
axes[1].set_ylabel('Frequency')

# This helps the two charts fit better without overlapping.
plt.tight_layout()

# This displays both subplots at the same time.
plt.show()
```
