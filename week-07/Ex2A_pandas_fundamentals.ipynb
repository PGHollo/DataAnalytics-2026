## Series

In this section, I am making small Python variables first, then turning them into pandas Series.

```python
# Import pandas so I can use Series and DataFrame
import pandas as pd
```

```python
# These are my four starting variables for the Series practice

aa = [1, 2, 3, 4]          # This is a simple list of numbers
ab = (5, 6, 7, 8)          # This is a tuple, which is kind of like a list that does not change
ac = [9, 10, 11, 12]       # This is another list of numbers
ad = {
    "first": 13,
    "second": 14,
    "third": 15,
    "fourth": 16
}                          # This dictionary uses labels instead of regular number positions
```

```python
# Now I am changing each variable into a pandas Series

ser1 = pd.Series(aa)   # This turns the aa list into a Series
ser2 = pd.Series(ab)   # This turns the ab tuple into a Series
ser3 = pd.Series(ac)   # This turns the ac list into a Series
ser4 = pd.Series(ad)   # This turns the ad dictionary into a Series

print("ser1:")
print(ser1)

print("\nser2:")
print(ser2)

print("\nser3:")
print(ser3)

print("\nser4:")
print(ser4)
```

```python
# This Series does not have 16 in it, so I want to see what pandas does with missing data

ser5 = pd.Series([17, 18, 19, 20])   # I made a Series with only four values

print(ser5)
print("\nThis Series only has the values I gave it. It does not include 16 because I did not add 16.")
```

## DataFrames from Series

Here, I am using the Series to build and update a DataFrame.

```python
# I am making a DataFrame from ser1
# A DataFrame is like a small table with rows and columns

df1 = pd.DataFrame(ser1)

df1
```

```python
# I am adding ser2 and ser3 as new columns using insert()
# The first number tells pandas where to place the new column

df1.insert(1, "ser2", ser2)   # Adds ser2 as the second column
df1.insert(2, "ser3", ser3)   # Adds ser3 as the third column

df1
```

```python
# Now I am adding ser4 as another column
# Since ser4 uses word labels, the index does not match df1's number index

df1["ser4"] = ser4

df1
```

### What happened when ser4 was added?

When I added `ser4`, pandas tried to line up the rows by the index labels.  
`df1` uses number indexes like `0, 1, 2, 3`, but `ser4` uses labels like `first`, `second`, `third`, and `fourth`.

Because the indexes do not match, pandas shows missing values as `NaN`.  
So basically, the data is there, but pandas does not know which row it should match with.

```python
# I am adding a new row using .loc
# Row 4 will have the values 5, 9, 12, and 0

df1.loc[4] = [5, 9, 12, 0]

df1
```

### Quick notes from `df2.info()`

## Coding a DataFrame

Now I am creating a DataFrame manually from a dictionary.

```python
# This dictionary is my sample student-style dataset
# Each key becomes a column in the DataFrame

test_data = {
    "ID": [1, 2, 3, 4, 5, 6, 7, 8],
    "Name": ["Alice", "Bob", "Charlie", "Dominic", "Eve", "Frank", "Grace", "Heidi"],
    "Age": [25, None, 22, 19, 31, 35, None, 18],
    "City": ["NY", "LA", "Chicago", "Houston", "Phoenix", "Boston", "Austin", "San Diego"],
    "Score": [85, 92, -1, 65, 78, None, 55, 90]
}
```

```python
# I am turning the dictionary into a DataFrame called df4

df4 = pd.DataFrame(test_data)

df4
```

```python
# I am exporting df4 into a CSV file
# index=False means pandas will not add an extra index column to the CSV

df4.to_csv("test_data.csv", index=False)

print("test_data.csv was created successfully.")
```
